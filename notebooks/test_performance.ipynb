```python
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "5"
os.environ["NVIDIA_VISIBLE_DEVICES"] = "5"

import pandas as pd
import emoji
import re

from transformers import AutoModelForSequenceClassification, AutoTokenizer
```

```python
def clean_text(text: str) -> str:

    text = emoji.replace_emoji(text, replace=' ')
    text = re.sub('@lentadnya', ' ', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\s+',' ', text)

    return text
```

```python
import sys 

sys.path.append('..')

model = AutoModelForSequenceClassification.from_pretrained("../logs/sentiment_experiments/ruroberta_runews_2class_512/checkpoint-9300").cuda().eval()
tokenizer = AutoTokenizer.from_pretrained("sberbank-ai/ruRoberta-large")
```

```python
df = pd.read_csv('../data/lentadnya_from_0_to_10000.csv').head(100)
```

```python
df.text = df.text.apply(lambda x: clean_text(x))
```

```python
import numpy as np
import torch
preds = []
for text in df.text.values.tolist():
    inp = tokenizer(text, return_tensors='pt')
    inp = inp.to(model.device)
    out = model(**inp)
    preds.append(np.argmax(torch.sigmoid(out[0]).detach().cpu().numpy(), axis=-1)[0])
```

```python
comparison_df = pd.read_excel('../data/test_sample.xlsx')
comparison_df['predicted_roberta'] = preds
```

```python
pd.set_option('display.expand_frame_repr', False)
comparison_df[comparison_df['predicted_roberta'] != comparison_df['predicted_label']]
```
